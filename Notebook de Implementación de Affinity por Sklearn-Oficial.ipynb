### Pre-procesamiento de datos

Antes de realizar el proceso de clustering por afinidad con la base de datos de un banco alemán, se debe hacer una extracción de la información a partir de las tres columnas relevantes a analizar: Edad, Monto y Duración de Crédito. 

```python
# Seleccionando las tres columnas de análisis: Edad, Monto y Duración
# de crédito de la base de datos
selected_cols = ["Age","Credit amount", "Duration"]
cluster_data = data.loc[:,selected_cols]
```

A continuación, se debe verificar el nivel de distribución de los datos para garantizar una segmentación lo más óptima posible y sin errores, mediante una gráfica estadística de barras y líneas de tendencias por cada uno de los tres atributos seleccionados.

```python
# Función para la creación de histogramas de distribución
def distributions(df):
    fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(8,8))
    sns.distplot(df["Age"], ax=ax1)
    sns.distplot(df["Credit amount"], ax=ax2)
    sns.distplot(df["Duration"], ax=ax3)
    plt.tight_layout()
```

```python
# Histograma de distribución de las tres columnas seleccionadas
distributions(cluster_data)
```

A partir de las gráficas, se puede observar un sesgo de las tres distribuciones con dirección a la izquierda, una asimetría que, si bien no es grave, no es apropiada para obtener los clusters más óptimos con la base de datos.

Para mejorar los resultados, se corrige este sesgo a partir de transformación logarítmica del subgrupo extraido anteriormente, obteniendo, mediante las gráficas como resultado, una distribución más centrada.

```python
# Normalización de segmentación entre conjuntos mediante una función de
# logaritmo natural en los datos a subdividir
cluster_log = np.log(cluster_data)
# Histograma de distribución de las tres columnas seleccionadas, con la
# normalización de base de datos
distributions(cluster_log)
```

Para finalizar el pre-procesamiento, se debe estandarizar los valores de los datasets de nuestra base de datos anteriormente normalizada para garantizar un mejor funcionamiento del código. Para esto, se utilizará la función StandardScaler de la librería Sklearn, para adecuar este proceso al sector de información requerido antes de transformarlo.

```python
# Uso de la función StandardScaler para establecer centros y escalamiento
# de variables en la base de datos normalizada
scaler = StandardScaler()
cluster_scaled = scaler.fit_transform(cluster_log)
```

### Ejecución del algoritmo de afinidad mediante Sklearn

En este algoritmo hay dos parámetros de entrada relevantes: la preferencia y el damping o similitud. Esto permite que el número de clusters no deba ser definido como un parámetro en el código, sino obtenido como resultado del proceso del método. 

Para verificar el número de subgrupos más optimo a analizar, se hará una ejecución previa de la función AffinityPropagation de la librería Sklearn con respecto a la base de datos pre-procesada, dándole como parámetros similitud 0.6 y un array de preferencias a revisar, todas en valor negativo para seguir los principios de limitación en las fórmulas del algoritmo; con los resultados siendo graficados para su mejor lectura.

```python
# Establecimiento de los valores de preferencia por columna de las tres escogidas
preferences = np.arange(-30,-190,-10)
clusters = []

# Invocación del método Affinity Propagation para determinar un conjunto de posibles números de clusters
# que se pueden generar para determinar la mejor preferencia a utilizar
for p in preferences:
    af = AffinityPropagation(preference=p, damping=0.6, max_iter=400, verbose=False).fit(cluster_scaled)
    labels_af = pd.DataFrame(af.labels_)
    clusters.append(len(af.cluster_centers_indices_))

# Gráfica de línea que indica, en un rango de preferencia -180 a -30, el número de
# clusters que se pueden generar con la información, para identificar la cifra
# más óptima de subgrupos según nuestros requerimientos
plt.figure(figsize=(10,7))
plt.xlabel("Preference")
plt.ylabel("Number of clusters")
plt.plot(preferences,clusters, marker='o')
```

Se puede observar que, en un rango inicial entre -180 y -140 de preferencia, la disminución del valor de esta tambien baja el número más óptimo de clusters a realizar; sin embargo, en valores posteriores, la cifra de grupos que se pueden crear aumenta. 

Para el ejemplo del uso del algoritmo de clustering por afinidad, se determinó que sería más óptimo un análisis de cuatro clusters, por lo que la preferencia a utilizar será de -140. 
Se invocará de manera definitiva a la función AffinityPropagation con nuestro dataset escalado, ingresando los parámetros mencionados; para luego asociar la información de la base de datos a los clusters generados por relaciones entre las tres columnas e ilustrarla en gráficas

```python
# Invocación oficial del algoritmo de Propagación de Afinidad con nuestra base de datos
# pre-procesada
af = AffinityPropagation(preference=-140, damping=0.6, verbose=False).fit(cluster_scaled)
# Toma del número de clusters y sus labels generados a partir de los resultados de la
# anterior función
labels_af = pd.DataFrame(af.labels_)
n_clusters_ = len(af.cluster_centers_indices_)

# Segmentación de la información de la base de datos a partir de los labels de clusters
# generados
clustered_data_af = cluster_data.assign(Cluster=labels_af)
scatters(clustered_data_af,'Cluster')

# Gráfica con la ilustración de los clusters generados por columnas a analizar
grouped_af = clustered_data_af.groupby(['Cluster']).mean().round(1)
```

Finalmente, para obtener un análisis e identificación más explicativa de cada uno de los clusters principales generados, se realiza un cálculo de la media de cada columna por subgrupo, para así identificar los siguientes detalles:
- Edad predominante
- Nivel de media del monto de crédito
- Duración promedio

```python
# Tabla de agrupamiento de los datos de las tres columnas por cluster, cada
# una por el uso de la media
grouped_af = clustered_data_af.groupby(['Cluster']).mean().round(1)
grouped_af
```
